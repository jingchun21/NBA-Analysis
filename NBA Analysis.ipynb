# Just focus on regular season and playoffs.

# Total home advantage in the past

```python
nba_games['wl_home'].value_counts().values[0]/nba_games.shape[0]
```

```python
nba_games.info()
```

# Change the type of game_date to datetime so can extract the year, month info

```python
nba_games['game_date'] = pd.to_datetime(nba_games['game_date'],format='%Y-%m-%d %H:%M:%S')
```

```python
nba_games.info()
```

```python
nba_games['game_year'] = nba_games['game_date'].dt.year
nba_games['game_month'] = nba_games['game_date'].dt.month
```

```python
nba_games.tail(20)
```

# Try to get the win and loss number in each year. 
Then create a table has win and loss number in each year and calculate the percentage of win at home.

```python
nba_wl_home = pd.DataFrame(nba_games.groupby('game_year')['wl_home'].value_counts())
```

```python
nba_wl_home.rename(columns = {'wl_home':'wl_home_counts'},inplace=True)
```

```python
nba_wl_home.head()
```

```python
nba_wl_home = nba_wl_home.reset_index().pivot(index='game_year',columns='wl_home',values='wl_home_counts').reset_index()
```

```python
nba_wl_home.head()
```

```python
nba_wl_home.columns
```

```python
nba_wl_home['PCT_Home_Win'] = round(nba_wl_home['W']/(nba_wl_home['W']+nba_wl_home['L'])*100,2)
```

```python
nba_wl_home.head()
```

```python
plt.plot(nba_wl_home['game_year'],nba_wl_home['PCT_Home_Win'])
```

# Results of home advantage from 1946-2023
In total, the home advantage is above 55% from 1946-2023. <br>
I am surprised to see the home advantage is going down. 

# NBA Total Score Growth

```python
nba_scores = nba_games[['game_year','pts_home','pts_away']]
```

```python
nba_scores.head()
```

```python
nba_scores['pts_total'] = nba_scores['pts_home']+nba_scores['pts_away']
```

```python
nba_scores.head()
```

```python
plt.plot(nba_scores['game_year'],nba_scores['pts_home'])
```

```python
plt.plot(nba_scores['game_year'],nba_scores['pts_total'])
```

```python
plt.plot(nba_scores['game_year'],nba_scores['pts_away'])
```

```python
nba_scores_total_year = nba_scores.groupby('game_year')['pts_total'].mean()
nba_scores_home_year = nba_scores.groupby('game_year')['pts_home'].mean()
nba_scores_away_year = nba_scores.groupby('game_year')['pts_away'].mean()
nba_scores_year = pd.concat([nba_scores_total_year,nba_scores_home_year,nba_scores_away_year],axis=1)
```

```python
nba_scores_year.head()
```

```python
nba_scores_year.plot()
plt.title('Total Scores from 1946-2023')
```

# Free Throw Analysis

```python
nba_ft = nba_games.loc[:,['game_year','ftm_home','fta_home','ft_pct_home','ftm_away','fta_away','ft_pct_away']]
```

```python
nba_ft.head()
```

```python
nba_ft.shape
```

```python
nba_ft.dropna()
```

```python
nba_ft['ft_pct_home']=round(nba_ft['ft_pct_home']*100,1)
nba_ft['ft_pct_away']=round(nba_ft['ft_pct_away']*100,1)
```

```python
nba_ftm_home_year = nba_ft.groupby('game_year')['ftm_home'].mean()
nba_fta_home_year = nba_ft.groupby('game_year')['fta_home'].mean()
nba_ft_pct_home_year = nba_ft.groupby('game_year')['ft_pct_home'].mean()
nba_ftm_away_year = nba_ft.groupby('game_year')['ftm_away'].mean()
nba_fta_away_year = nba_ft.groupby('game_year')['fta_away'].mean()
nba_ft_pct_away_year = nba_ft.groupby('game_year')['ft_pct_away'].mean()
nba_ft_year = pd.concat([nba_ftm_home_year,nba_fta_home_year,nba_ft_pct_home_year,nba_ftm_away_year,nba_fta_away_year,nba_ft_pct_away_year],axis=1)
```

```python
nba_ft_year.head()
```

```python
nba_ft_year.plot(figsize=(8,12))
plt.title('Free Throws from 1946-2023')
```

Home games have more free throws. 
<br> The percentage of successful free throw is similar home vs away.
